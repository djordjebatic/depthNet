```python
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

imgL_o = Image.open("data_scene_flow/training/disp_occ_0/000092_10.png")

arr = np.asarray(imgL_o)
plt.imshow(arr)
print('min: %f, max: %f, mean: %f' % (np.min(arr), np.max(arr), np.mean(arr)))
```

```python
data = np.ascontiguousarray(imgL_o, dtype=np.float32)/256
#arr = np.asarray(data)
plt.imshow(data < 50)
img = (data*256).astype('uint16')
img = Image.fromarray(img)
img.save("nest.png")
print('min: %f, max: %f, mean: %f' % (np.min(data), np.max(data), np.mean(data)))
```
